```python
data_ware_drops = data_ware.drop(['Expected Ship Date', 'Order Priority','Consignee Zip', 'Consignee State', 'ORD_ZIPCODE' ], axis = 1)
```

```python
data_ware_drops.head(10)
```

```python
data_ware_drops["Wave Number"] = data_ware_drops['Wave Number'].str.replace('[^\w\s]','')
data_ware_drops["Order Number"] = data_ware_drops['Order Number'].str.replace('[^\w\s]','')
data_ware_drops["WEIGHT"] = data_ware_drops['WEIGHT'].str.replace('[^\w\s]','')
data_ware_drops["QTY_PER_LINES"] = data_ware_drops['QTY_PER_LINES'].str.replace('[^\w\s]','')
data_ware_drops["Order Type"] = data_ware_drops['Order Type'].str.replace('[^\w\s]','')
```

```python
data_ware_drops.head()
```

```python
New_data = data_ware_drops.loc[:, ['Wave Number','Order Number','Order Line Number','Order Number.1','Destination Location','Consignee City','Number of Lines per Order','QTY_PER_LINES', 'WEIGHT','Consignee','Order Type']]
```

```python
New_data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
```

```python
New_data.iloc[:, 4] = labelencoder.fit_transform(New_data.iloc[:, 4])
```

```python
New_data.iloc[:, 5] = labelencoder.fit_transform(New_data.iloc[:, 5])
New_data.iloc[:, 6] = labelencoder.fit_transform(New_data.iloc[:, 6])
```

```python
X = New_data.iloc[:,:-2]
y = New_data.iloc[:,-1]
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
```

```python
predicted_labels = rfc.predict(X_test)
```

```python
predicted_labels
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predicted_labels))
```

```python
## Model works accuracy with 98%
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cf = confusion_matrix(y_test, predicted_labels)
```
